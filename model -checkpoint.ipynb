# Feature Engineering

```python
# Skills preprocessing
import ast

# 1. Handle missing values
def clean_skills(skills):
    """Convert skills to a clean list format"""
    if pd.isna(skills):
        return []
    
    # Handle string representations of lists
    if isinstance(skills, str) and skills.startswith('['):
        try:
            skills = ast.literal_eval(skills)
        except:
            skills = [skills]
    
    # Ensure we have a list
    if not isinstance(skills, list):
        skills = [skills]
    
    # Clean each skill
    cleaned = []
    for skill in skills:
        if isinstance(skill, list):
            cleaned.extend(clean_skills(skill))
        elif pd.notna(skill):
            skill_str = str(skill).strip().lower()
            if skill_str:
                cleaned.append(skill_str)
    
    return list(set(cleaned))  # Remove duplicates

# Apply cleaning to skills column
df['cleaned_skills'] = df['skills'].apply(clean_skills)

# 2. Get top skills
all_skills = [skill for skills in df['cleaned_skills'] for skill in skills]
top_skills = pd.Series(all_skills).value_counts().head(50).index.tolist()

# 3. Prepare for MultiLabelBinarizer
def filter_top_skills(skills_list):
    """Keep only skills that are in our top skills list"""
    return [skill for skill in skills_list if skill in top_skills]

df['filtered_skills'] = df['cleaned_skills'].apply(filter_top_skills)

# 4. Binarize skills
mlb = MultiLabelBinarizer(classes=top_skills)
skills_matrix = mlb.fit_transform(df['filtered_skills'])
skills_df = pd.DataFrame(skills_matrix, columns=mlb.classes_, index=df.index)

# 5. Merge back with original data
final_df = pd.concat([df, skills_df], axis=1)

```

```python
def clean_skill(skill):
    """Normalize a single skill string"""
    if pd.isna(skill):
        return None
    
    # Convert to string and standardize
    skill = str(skill).strip().lower()
    
    # Remove special chars but keep spaces/hyphens
    skill = re.sub(r'[^\w\s-]', '', skill)
    
    # Replace common variants
    skill = re.sub(r'\bjs\b|\bjavascript\b', 'javascript', skill)
    skill = re.sub(r'\bml\b|\bmachinelearning\b', 'machine learning', skill)
    skill = re.sub(r'\bdl\b|\bdeeplearning\b', 'deep learning', skill)
    skill = re.sub(r'\bnlp\b', 'natural language processing', skill)
    
    return skill if skill else None

def flatten_skills(skill_list):
    """Convert mixed formats (strings, lists) to clean list"""
    if pd.isna(skill_list):
        return []
    
    # Handle string representations of lists
    if isinstance(skill_list, str) and skill_list.startswith('['):
        try:
            skill_list = eval(skill_list)  # Safe for trusted data
        except:
            skill_list = [skill_list]
    
    # Ensure we have an iterable
    if not isinstance(skill_list, (list, tuple)):
        skill_list = [skill_list]
    
    # Clean each skill
    cleaned = []
    for item in skill_list:
        cleaned_skill = clean_skill(item)
        if cleaned_skill:
            cleaned.append(cleaned_skill)
    
    return list(set(cleaned))  # Remove duplicates

# Apply to entire column
df['cleaned_skills'] = df['skills'].apply(flatten_skills)
```

```python
# Manual mapping for similar skills
skill_synonyms = {
    'python': ['python', 'python3', 'python programming'],
    'machine learning': ['ml', 'machinelearning'],
    'sql': ['mysql', 'postgresql', 'sql server'],
    'aws': ['amazon web services']
}

def consolidate_skills(skill_list):
    consolidated = []
    for skill in skill_list:
        # Check if skill matches any synonym group
        matched = False
        for standard, variants in skill_synonyms.items():
            if skill in variants or skill == standard:
                consolidated.append(standard)
                matched = True
                break
        if not matched:
            consolidated.append(skill)
    return list(set(consolidated))

df['consolidated_skills'] = df['cleaned_skills'].apply(consolidate_skills)
```

```python
# Get frequency of all skills
all_skills = [skill for sublist in df['consolidated_skills'] for skill in sublist]
skill_counts = pd.Series(all_skills).value_counts()

# Select top N skills (adjust as needed)
top_skills = skill_counts.head(50).index.tolist()

# Filter to only include top skills
df['top_skills'] = df['consolidated_skills'].apply(
    lambda x: [skill for skill in x if skill in top_skills]
)
```

```python

# 1. Load the data
data = pd.read_csv('resume_data_processed.csv')
print("Available columns:", data.columns.tolist())

# 2. Feature Engineering (using your skills preprocessing)
X = data.filter(regex='^skill_').values
print(f"Found {X.shape[1]} skill features")

# 3. Target Selection - Multiple Approaches
def prepare_targets(df):
    """Flexible target column preparation"""
    # Approach 1: Check for standard column names
    for col in ['job_title_id', 'job_title', 'role', 'position']:
        if col in df.columns:
            return df[col], 'classification'
    
    # Approach 2: Check for salary data
    for col in ['salary_normalized', 'salary', 'compensation']:
        if col in df.columns:
            return df[col], 'regression'
    
    # Approach 3: Create synthetic targets from skills
    if len(df.filter(regex='^skill_').columns) > 0:
        print("Creating skill-based clusters as synthetic targets")
        from sklearn.cluster import KMeans
        kmeans = KMeans(n_clusters=5, random_state=42)
        clusters = kmeans.fit_predict(df.filter(regex='^skill_'))
        return clusters, 'clustering'
    
    # If all else fails
    raise ValueError("No usable target columns found and cannot create synthetic targets")

y, problem_type = prepare_targets(data)
print(f"Using target type: {problem_type}")

# 4. Train/Test Split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2, 
    stratify=y if problem_type in ['classification', 'clustering'] else None,
    random_state=42
)

print("\nData ready for modeling:")
print(f"- Training samples: {X_train.shape[0]}")
print(f"- Test samples: {X_test.shape[0]}")
print(f"- Problem type: {problem_type}")
```

```python


# 1. Load and verify data
try:
    data = pd.read_csv('resume_data_processed.csv')
    print("Data loaded successfully. Columns:", data.columns.tolist())
except Exception as e:
    print("Error loading data:", e)
    raise

# 2. Prepare skill features
skill_cols = [col for col in data.columns if col.startswith('skill_')]
if not skill_cols:
    raise ValueError("No skill columns found (expected columns starting with 'skill_')")

X = data[skill_cols].values
print(f"Using {len(skill_cols)} skill features")

# 3. Dimensionality reduction
pca = PCA(n_components=2)  # Using 2D for simpler visualization
X_pca = pca.fit_transform(X)

# 4. Cluster optimization
best_k = 3  # Default fallback
try:
    silhouette_scores = []
    for k in range(2, min(10, len(X)-1)):  # Ensure k < n_samples
        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        labels = kmeans.fit_predict(X)
        silhouette_scores.append(silhouette_score(X, labels))
    
    best_k = np.argmax(silhouette_scores) + 2
    print(f"Optimal cluster count: {best_k} (scores: {silhouette_scores})")
except Exception as e:
    print("Using default cluster count due to error:", e)

# 5. Final clustering
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
data['cluster'] = kmeans.fit_predict(X)

# 6. Visualization
plt.figure(figsize=(12, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=data['cluster'], cmap='viridis', s=50)
plt.colorbar(scatter, label='Cluster')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title(f'Resume Skill Clusters (k={best_k})')

# Add cluster centers
centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers_pca[:, 0], centers_pca[:, 1], c='red', s=200, alpha=0.5, marker='X')
plt.grid(True)
plt.show()

# 7. Cluster analysis
print("\nCluster Profiles:")
for cluster in sorted(data['cluster'].unique()):
    cluster_size = sum(data['cluster'] == cluster)
    print(f"\nCluster {cluster} ({cluster_size} resumes)")
    
    # Get top skills
    top_skills = data[data['cluster'] == cluster][skill_cols].mean().sort_values(ascending=False).head(5)
    print("Top skills:")
    for skill, score in top_skills.items():
        print(f"- {skill[6:]}: {score:.2f}")  # Remove 'skill_' prefix
    
    # Sample positions
    positions = data[data['cluster'] == cluster]['positions'].dropna().unique()[:3]
    print("Example positions:", ", ".join(positions[:3]))

# 8. Save results
data.to_csv('clustered_resumes.csv', index=False)
print("\nResults saved to 'clustered_resumes.csv'")
# Add this after clustering to evaluate quality
print(f"Silhouette Score: {silhouette_score(X, data['cluster']):.2f}")
print("Cluster distribution:\n", data['cluster'].value_counts())
```

```python
def recommend_skills(resume_id, n=5):
    """Suggest complementary skills from the same cluster"""
    cluster = data.loc[resume_id, 'cluster']
    cluster_skills = data[data['cluster'] == cluster][skill_cols].mean()
    current_skills = data.loc[resume_id, skill_cols]
    
    # Recommend skills not already present
    recommendations = cluster_skills[current_skills == 0].sort_values(ascending=False).head(n)
    return recommendations.index.str.replace('skill_', '').tolist()

# Example usage
print("Recommended skills for resume 42:", recommend_skills(42))
```

```python
# Compare clusters to job requirements
def analyze_skill_gaps(target_cluster, job_requirements):
    """job_requirements: List of required skills"""
    cluster_skills = data[data['cluster'] == target_cluster][skill_cols].mean()
    gaps = {s: cluster_skills[f'skill_{s}'] for s in job_requirements 
            if f'skill_{s}' in skill_cols}
    return pd.Series(gaps).sort_values()

# Example
print(analyze_skill_gaps(0, ['python', 'tensorflow', 'sql', 'tableau']))
```

```python
from sklearn.metrics.pairwise import cosine_similarity

def find_similar_resumes(resume_id, n=3):
    """Find most similar resumes within cluster"""
    cluster = data.loc[resume_id, 'cluster']
    cluster_data = data[data['cluster'] == cluster]
    
    similarities = cosine_similarity(
        [cluster_data.loc[resume_id, skill_cols]],
        cluster_data[skill_cols]
    )[0]
    
    most_similar = np.argsort(-similarities)[1:n+1]  # Exclude self
    return cluster_data.iloc[most_similar][['positions', 'skills']]

# Example
print("Similar resumes to #42:")
find_similar_resumes(42)
```

```python
import json
import joblib
import numpy as np
from datetime import datetime

# 1. Convert numpy types to native Python types for JSON serialization
def convert_numpy(obj):
    if isinstance(obj, (np.int32, np.int64)):
        return int(obj)
    elif isinstance(obj, np.float32):
        return float(obj)
    elif isinstance(obj, np.ndarray):
        return obj.tolist()
    elif isinstance(obj, dict):
        return {k: convert_numpy(v) for k, v in obj.items()}
    elif isinstance(obj, list):
        return [convert_numpy(item) for item in obj]
    return obj

# 2. Prepare metadata with type conversion
metadata = {
    'n_clusters': int(best_k),  # Explicit conversion to native Python int
    'features_used': int(len(skill_cols)),
    'creation_date': datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    'silhouette_score': float(silhouette_score(X, data['cluster'])),  # Convert to float
    'cluster_names': {int(k): str(v) for k, v in cluster_names.items()}  # Ensure keys/values are serializable
}

# 3. Save all artifacts
model_dir = "deployment_artifacts"
os.makedirs(model_dir, exist_ok=True)

# Save binary objects
joblib.dump({
    'model': kmeans,
    'skill_columns': skill_cols,
    'pca': pca,
    'cluster_descriptions': cluster_names
}, f"{model_dir}/model_artifacts.pkl")

# Save human-readable metadata
with open(f"{model_dir}/metadata.json", 'w') as f:
    json.dump(convert_numpy(metadata), f, indent=2)  # Apply type conversion

print(f"""
✅ Model saved successfully in {model_dir}/
├── model_artifacts.pkl  # All binary objects
└── metadata.json       # Human-readable info

Model Details:
- Clusters: {metadata['n_clusters']}
- Features: {metadata['features_used']}
- Silhouette Score: {metadata['silhouette_score']:.2f}
- Top Skills per Cluster:
""")

# Print cluster profiles
for cluster_id, name in metadata['cluster_names'].items():
    top_skills = (
        data[data['cluster'] == cluster_id][skill_cols]
        .mean()
        .sort_values(ascending=False)
        .head(3)
        .index.str.replace('skill_', '')
        .tolist()
    )
    print(f"  {cluster_id}: {name} → {', '.join(top_skills)}")
```

```python
import joblib
import json

# Load artifacts
artifacts = joblib.load('deployment_artifacts/model_artifacts.pkl')
kmeans = artifacts['model']
skill_cols = artifacts['skill_columns']

# Load metadata
with open('deployment_artifacts/metadata.json') as f:
    metadata = json.load(f)

print(f"Loaded {metadata['n_clusters']} cluster model (score: {metadata['silhouette_score']:.2f})")

# Example prediction
def predict_cluster(skills_list):
    """Predict cluster for new resume"""
    features = np.zeros(len(skill_cols))
    for skill in skills_list:
        skill_col = f"skill_{skill.lower().replace(' ', '_')}"
        if skill_col in skill_cols:
            features[skill_cols.index(skill_col)] = 1
    return kmeans.predict([features])[0]

# Test
print("Example prediction:", predict_cluster(['python', 'machine learning']))
```
